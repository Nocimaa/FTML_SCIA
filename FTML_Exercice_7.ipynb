### Problématique

Comment identifier et caractériser les différents segments de marché dans le secteur de la location d’appartements afin d’optimiser le positionnement tarifaire et l’expérience utilisateur ?

### Contexte

- Les plateformes immobilières listent des milliers d’annonces aux profils très variés (studios, T2, lofts, etc.).  
- Les locataires et les propriétaires peinent à se repérer dans cette offre hétérogène.

### Approche proposée

1. **Sélection des variables clés**  
   - Quantitatives : loyer, surface, nombre de chambres et salles de bain, coordonnées géographiques.  
   - Qualitatives : équipements (AC, parking, balcon…) convertis en variables binaires.

2. **Regroupement par clustering**  
   - Application d’algorithmes non supervisés (K-means, DBSCAN, hiérarchique).  
   - Détection de « sous-marchés » naturels (ex. studios premium vs. T2 économiques).

### Résultats attendus

- **Pour les propriétaires** : ajustement fin des loyers selon chaque segment de clientèle.  
- **Pour les plateformes** : filtres et recommandations plus pertinentes.  
- **Pour les locataires** : recherche ciblée sur le type d’appartement qui correspond à leurs critères et leur budget.

## 2. Standardisation des features

Avant d'exécuter un algorithme de clustering, nous standardisons les features afin qu'elles aient une moyenne nulle et une variance unitaire. Cela garantit qu'aucune caractéristique ne domine la mesure de la distance euclidienne.

```python
X_num = X_train.select_dtypes(include=[np.number]).dropna()
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_num)
X_scaled_df = pd.DataFrame(X_scaled, columns=num_cols)
X_scaled_df.describe().T
```

## Clustering par K-means

Nous exécutons la méthode K-Means pour des tailles allant de 2 à 8. Pour chaque \(k\), nous enregistrons :  
- **Inertie** (somme des carrés intra-cluster) : plus la valeur est faible, meilleure est la compacité, mais elle diminue toujours lorsque \(k\) augmente.  
- **Score Silhouette** (utilisant les distances euclidiennes sur les données standardisées 13D) : plus la valeur est élevée, meilleure est la séparation des clusters.

---

### Pourquoi utiliser l’inertie ?

- **Mesure de cohésion interne** : l’inertie quantifie la distance au carré de chaque point à son centroïde. Une inertie faible signifie que les points sont bien agrégés autour de leur centre.  
- **Méthode du coude** : en traçant l’inertie en fonction de \(k\), on repère souvent un “coude” — le point où la diminution d’inertie se ralentit. C’est un bon compromis entre complexité (plus de clusters) et gain de compacité.  
- **Simplicité de calcul** : rapide à obtenir à chaque itération de K-Means, elle sert de critère d’optimisation interne de l’algorithme.

### Pourquoi utiliser le score de silhouette ?

- **Combinaison de cohésion et de séparation** : pour chaque point \(x\), on calcule  
  - \(a(x)\) : distance moyenne aux points de son cluster  
  - \(b(x)\) : distance moyenne au cluster le plus proche  
  Le score silhouette $s(x) = \frac{b(x) - a(x)}{\max(a(x), b(x))}$
 varie entre –1 et +1.  
- **Évaluation globale normalisée** : la moyenne des \(s(x)\) permet de comparer directement des partitions avec des valeurs de \(k\) différentes, sans que le score augmente artificiellement quand \(k\) croît.  
- **Détection des points mal classés** : un \(s(x)\) négatif signale un point plus proche d’un autre cluster que du sien, pointant d’éventuels problèmes d’assignation.

---

### Complémentarité

1. **Inertie** guide le choix initial de \(k\) via la méthode du coude, en se concentrant sur la compacité interne.  
2. **Silhouette** valide ensuite la qualité du partitionnement en prenant en compte la séparation entre clusters et en détectant les anomalies.  

En combinant ces deux métriques, on obtient à la fois un bon compromis pour déterminer le nombre de clusters et une assurance de la qualité du clustering produit.


```python
X_num = X_train.select_dtypes(include=[np.number]).copy()
X_num = X_num.dropna()

ks = range(2, 9)
inertias = []
silhouettes = []

for k in ks:
    kmeans = KMeans(n_clusters=k, n_init=20, random_state=42)
    labels = kmeans.fit_predict(X_scaled)
    inertias.append(kmeans.inertia_)
    silhouettes.append(silhouette_score(X_scaled, labels))
    print(f"k = {k:<2}  inertia = {kmeans.inertia_:>8.1f}  silhouette = {silhouettes[-1]:.3f}")
```

```python
# Plot Inertia & Silhouette vs. k side by side
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(ks, inertias, marker='o')
plt.xticks(ks)
plt.xlabel("Number of clusters (k)")
plt.ylabel("Inertia (WCSS)")
plt.title("K-Means: Elbow Plot")

plt.subplot(1, 2, 2)
plt.plot(ks, silhouettes, marker='o', color='darkorange')
plt.xticks(ks)
plt.xlabel("Number of clusters (k)")
plt.ylabel("Average Silhouette Score")
plt.title("K-Means: Silhouette vs. k")

plt.tight_layout()
plt.show()
```

## Clustering par Gaussian mixture models

Pour évaluer la qualité et sélectionner le nombre optimal de composantes \( k \) pour notre modèle de mélange gaussien, nous utilisons trois métriques complémentaires :

- **AIC (Akaike Information Criterion)**
  - Mesure le compromis entre la qualité de l'ajustement du modèle et sa complexité (nombre de paramètres).
  - Un AIC plus faible indique un meilleur compromis (meilleure qualité de fit pour un nombre de paramètres donné).
  - Utile pour comparer des modèles avec des nombres de composantes différents, car il pénalise les modèles trop complexes.

- **BIC (Bayesian Information Criterion)**
  - Similaire à l'AIC mais pénalise davantage la complexité du modèle, favorisant des modèles plus parsimonieux.
  - Un BIC plus faible indique également un meilleur modèle.
  - Plus strict que l'AIC, il est recommandé lorsque l'on craint le sur-apprentissage.

- **Silhouette Score**
  - Mesure la cohésion interne et la séparation inter-clusters.
  - Défini pour chaque point, il varie de -1 à +1 :
    - Valeur proche de +1 : point bien assigné à son cluster et éloigné des autres clusters.
    - Valeur proche de 0 : point à la frontière de deux clusters.
    - Valeur négative : point mal assigné.
  - La moyenne sur tous les points permet d'évaluer globalement la qualité des clusters.

Ces métriques sont choisies car :
- AIC et BIC permettent de sélectionner le nombre de composantes de manière statistique, en prenant en compte la complexité du modèle pour éviter le sur-apprentissage.
- Le Silhouette Score complète ces critères « d’information » en jugeant qualitativement la structure des clusters obtenus.
- Leur combinaison garantit une sélection robuste, à la fois basée sur la vraisemblance du modèle et sur la forme des clusters.


```python
aics, bics, silhouettes_gmm = [], [], []

for k in ks:
    gmm = GaussianMixture(
        n_components=k,
        covariance_type='full',
        n_init=10,
        random_state=42
    )
    gmm.fit(X_scaled)
    labels_gmm = gmm.predict(X_scaled)

    aics.append(gmm.aic(X_scaled))
    bics.append(gmm.bic(X_scaled))

    silhouettes_gmm.append(silhouette_score(X_scaled, labels_gmm))

    print(f"k = {k:<2}  AIC = {aics[-1]:>10.1f}  "
          f"BIC = {bics[-1]:>10.1f}  "
          f"silhouette = {silhouettes_gmm[-1]:.3f}")
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(ks, aics, marker='o', label='AIC')
plt.plot(ks, bics, marker='o', label='BIC')
plt.xlabel("Nombre de composantes (k)")
plt.ylabel("Critère d’information")
plt.title("AIC et BIC en fonction de k")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(ks, silhouettes_gmm, marker='o', color='C2')
plt.xlabel("Nombre de composantes (k)")
plt.ylabel("Silhouette")
plt.title("Silhouette GMM vs k")

plt.tight_layout()
```

## Comparaison K-Means vs GMM

- **K-Means**  
  - Score de silhouette maximal à *k = 3* (≈ 0.41), puis décroissance progressive au-delà.  
  - Inertie (WCSS) chute fortement jusqu’à k = 5, puis se stabilise (coude entre 3 et 5 clusters).

- **GMM (Gaussian Mixture Models)**  
  - Score de silhouette maximal à *k = 3* (≈ 0.37), mais valeur nettement inférieure à K-Means.  
  - Critères AIC/BIC diminuent jusqu’à k = 5, suggérant un compromis complexité/fit.

> **Choix selon vos priorités** :  
> - Séparation nette et simplicité → **K-Means**  
> - Modélisation probabiliste et flexibilité → **GMM**

```python
ks = [2, 3, 4, 5, 6, 7, 8]

# Scores de silhouette mesurés pour chaque k
sil_kmeans = [0.363, 0.410, 0.368, 0.297, 0.309, 0.322, 0.303]
sil_gmm    = [0.327, 0.372, 0.121, 0.125, 0.172, 0.135, 0.131]

plt.figure(figsize=(8, 5))
plt.plot(ks, sil_kmeans, marker='o', label='K-Means')
plt.plot(ks, sil_gmm,    marker='x', label='GMM')
plt.xticks(ks)
plt.xlabel('Nombre de clusters/composantes (k)')
plt.ylabel('Score de silhouette')
plt.title('Comparaison des scores de silhouette pour K-Means et GMM')
plt.legend()
plt.grid(linestyle='--', alpha=0.5)
plt.show()
```
