```python
# FTML 2025 - Question 8 de la partie 3

import numpy as np
import matplotlib.pyplot as plt

n = 100 
d = 5 
sigma2_true = 2.0

np.random.seed(69)

X = np.random.randn(n, d)

theta_star = np.random.randn(d)

epsilon = np.random.randn(n) * np.sqrt(sigma2_true)

y = X @ theta_star + epsilon

XtX_inv = np.linalg.inv(X.T @ X)
theta_hat = XtX_inv @ X.T @ y

residuals = y - X @ theta_hat
sigma2_hat = (residuals.T @ residuals) / (n - d)

print(f"Sigma² estimé : {sigma2_hat:.4f}")
print(f"Sigma² théorique : {sigma2_true:.4f}")

plt.hist(residuals, bins=30, density=True, alpha=0.6)
plt.title("Distribution des résidus")
plt.xlabel("Résidus")
plt.ylabel("Densité")
plt.show()
```
